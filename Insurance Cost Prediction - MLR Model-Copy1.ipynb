# TRAINING A LINEAR REGRESSION MODEL

# Seperating the Target from DataFrame

```python
# First we will split our data into â€˜xâ€™ array that contains the features and a â€˜yâ€™ array with the target variable
```

```python
# Data Preparation
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
print("X:\n",x)
print()
print("Y:\n",y)
```

# Splitting the Dataset into Training and Testing

```python
# Next we will split our dataset (insurance.csv) into a training set and a testing set. 
# We will train our model on the training set and then use the test set to evaluate the model.
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25, random_state=2)
print("X Train: ",x_train.shape)
print("X Test:  ",x_test.shape)
print("Y Train: ",y_train.shape)
print("Y Test:  ",y_test.shape)
print(x_train)
```

```python
# â€˜test_sizeâ€™ represents that 30% of the data goes to the test data and the rest to the training set.
# Setting random state as an integer, train_test_split will returnÂ sameÂ results for each execution.
```

# Building a MLR Model

```python
# Building the Linear Regression Model
lr=LinearRegression()
# Fit linear model by passing training dataset
lr.fit(x_train,y_train)
```

# P-Values
